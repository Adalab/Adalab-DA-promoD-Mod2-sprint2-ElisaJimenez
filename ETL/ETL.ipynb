```python

def extraer_datos(urls):
    
    # metemos las 3 url en una lista par poder iterar sobre ella
    urls = ["http://universities.hipolabs.com/search?country=canada",
    "http://universities.hipolabs.com/search?country=argentina",
    "http://universities.hipolabs.com/search?country=united states"]

    union_paises = []

    #iteramos por la lista para extraer los datos de las 3 url
    for url in urls:
        response = requests.get(url)
        datos_uni_paises = response.json()
        df = pd.DataFrame(datos_uni_paises)
        union_paises.append(df)

    #concatenamos los df
    df_paises = pd.concat(union_paises)
        
    return df_paises
```
